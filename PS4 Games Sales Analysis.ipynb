```python
data = {
    'Genre': ['Action', 'Action-Adventure', 'Adventure', 'Fighting', 'MMO', 'Misc', 'Music', 'Party', 
              'Platform', 'Puzzle', 'Racing', 'Role-Playing', 'Shooter', 'Simulation', 'Sports', 
              'Strategy', 'Visual Novel'],
    'North America': [48.15, 22.30, 5.27, 8.58, 1.32, 3.64, 2.40, 0.09, 5.48, 0.28, 5.86, 22.15, 50.50, 1.47, 33.31, 0.57, 0.20],
    'Europe': [57.51, 27.79, 6.08, 6.11, 1.45, 6.03, 1.62, 0.47, 8.92, 0.16, 14.77, 22.40, 58.64, 2.07, 42.81, 0.22, 0.02],
    'Japan': [10.27, 1.96, 1.64, 1.67, 0.20, 0.96, 0.17, 0.00, 0.67, 0.00, 0.76, 9.33, 4.42, 0.28, 1.83, 0.34, 0.19],
    'Rest of World': [20.76, 9.83, 2.23, 2.97, 0.55, 1.85, 0.84, 0.10, 2.77, 0.10, 3.81, 8.91, 21.48, 0.71, 14.90, 0.17, 0.05]
}

df = pd.DataFrame(data)

df_long = df.melt(id_vars="Genre", var_name="Region", value_name="Sales")

plt.figure(figsize=(12, 8))

sns.barplot(x='Genre', y='Sales', hue='Region', data=df_long)

plt.title('Regional Sales by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Sales (Millions)', fontsize=12)

plt.xticks(rotation=90)

plt.tight_layout()

plt.show()
```

```python
# 3. Find the publisher with the highest total global sales. 
publisher_sales = df.groupby('Publisher')['Global'].sum().reset_index()
top_publisher = publisher_sales.sort_values(by='Global', ascending=False).iloc[0]
print(f"Publisher with the highest global sales: {top_publisher['Publisher']}")
print(f"Total Global Sales: {top_publisher['Global']}")
```

```python
# 4. Analyze the year with the highest number of game releases.
release_count_per_year = df['Year'].astype(int).value_counts().sort_index()
release_count_per_year
```

```python
max_release_year = release_count_per_year.idxmax()
max_release_count = release_count_per_year.max()
print(f"The year with the highest number of game releases is {max_release_year} with {max_release_count} releases.")
```

```python
plt.figure(figsize=(10, 6))
release_count_per_year.plot(kind='bar', color='blue')
plt.title('Number of Game Releases per Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=90)
plt.show()
```

```python
# 5. Determine the most popular genre globally based on total sales
genre_sales = df.groupby('Genre')['Global'].sum().reset_index()
most_popular_genre = genre_sales.sort_values(by='Global',ascending=False).iloc[0]
most_popular_genre
```

```python
japan_sales = df.groupby('Game')['Japan'].sum().reset_index()

japan_sales_sorted = japan_sales.sort_values(by='Japan', ascending=False)

game_highest_sales_japan = japan_sales_sorted.iloc[0]

print(f"Game with Highest Sales in Japan: {game_highest_sales_japan['Game']}")
print(f"Sales in Japan: {game_highest_sales_japan['Japan']}")
```

```python
# 7. Examine if there is a correlation between sales in North America and Europe.
correlation = df['North America'].corr(df['Europe'])
print(f"Correlation between North America and Europe sales: {correlation}")
```

```python
# 8. Compare the sales growth trend for "Action" and "Shooter" genres over the years.

df['Year'] = df['Year'].astype(int)

df_filtered = df[df['Genre'].isin(['Action', 'Shooter'])]
genre_sales_trends = df_filtered.groupby(['Year', 'Genre'])['Global'].sum().reset_index()
print(genre_sales_trends)
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(data=genre_sales_trends, x='Year', y='Global', hue='Genre', palette='Set2')

plt.title('Total Global Sales for "Action" and "Shooter" Genres Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Global Sales (millions)', fontsize=14)

plt.legend(title='Genre')
plt.tight_layout()

plt.show()
```

```python
# 9. Identify the top publisher in terms of sales for each region.
regions = ['North America', 'Europe', 'Japan', 'Rest of World']

top_publishers_per_region = []

for region in regions:
    publisher_sales = df.groupby('Publisher')[region].sum().reset_index()
    top_publisher = publisher_sales.loc[publisher_sales[region].idxmax()]
    top_publishers_per_region.append({
        'Region': region,
        'Top Publisher': top_publisher['Publisher'],
        'Sales': top_publisher[region]
    })

top_publishers_df = pd.DataFrame(top_publishers_per_region)
top_publishers_df
```

```python
top_publishers_df = pd.DataFrame({
    'Region': ['North America', 'Europe', 'Japan', 'Rest of World'],
    'Top Publisher': ['Activision', 'Activision', 'Square Enix', 'Activision'],
    'Sales': [28.20, 30.96, 5.95, 11.71]
})

colors = ['lightblue', 'lightgreen', 'salmon', 'orange']

plt.figure(figsize=(10, 6))
bars = plt.bar(top_publishers_df['Region'], top_publishers_df['Sales'], color=colors)

for i, row in top_publishers_df.iterrows():
    plt.text(i, row['Sales'] + 0.5, row['Top Publisher'], ha='center', color='black', fontsize=10)

plt.xlabel('Region', fontsize=12)
plt.ylabel('Sales (Millions)', fontsize=12)
plt.title('Top Publisher by Sales in Each Region', fontsize=14)

plt.subplots_adjust(top=0.85, bottom=0.15)

plt.tight_layout()
plt.show()
```

```python
# Total sales comparison for regions (bar chart).

region_sales = df[['North America', 'Europe', 'Japan', 'Rest of World']].sum()

region_sales_df = region_sales.reset_index()
region_sales_df.columns = ['Region', 'Total Sales (Millions)']

plt.figure(figsize=(10, 6))

sns.barplot(x='Region', y='Total Sales (Millions)', data=region_sales_df)

plt.title('Total Sales Comparison Across Regions', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Sales (Millions)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Share of total global sales by genre (pie chart).

genre_sales = df.groupby('Genre')['Global'].sum()


total_global_sales = genre_sales.sum()


genre_sales_percentage = (genre_sales / total_global_sales) * 100

plt.figure(figsize=(8, 8))

wedges, texts, autotexts = plt.pie(
    genre_sales_percentage, 
    labels=None,
    autopct='%1.1f%%',
    startangle=90, 
    colors=plt.cm.Paired.colors,
    pctdistance=1.2,
    labeldistance=1.5,
)

for autotext in autotexts:
    autotext.set_fontsize(7)

plt.legend(
    labels=genre_sales_percentage.index,
    title='Genres',  
    loc='center left', 
    bbox_to_anchor=(1.05, 0.5), 
    fontsize=12
)

plt.title('Share of Total Global Sales by Genre', fontsize=16)

plt.axis('equal')

plt.tight_layout()
plt.show()
```

```python
# Trend of global sales over the years (line chart).

global_sales_by_year = df.groupby('Year')['Global'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(global_sales_by_year['Year'], global_sales_by_year['Global'], marker='o', color='b', linestyle='-', linewidth=2, markersize=6)
plt.title('Trend of Global Sales Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Global Sales (Millions)', fontsize=12)

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Top 10 best-selling games globally (horizontal bar chart).

top_10_selling_games = df.sort_values(by='Global', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_10_selling_games['Game'], top_10_selling_games['Global'], color='skyblue')

plt.title('Top 10 Best-Selling Games Globally', fontsize=16)
plt.xlabel('Global Sales (Millions)', fontsize=12)
plt.ylabel('Game', fontsize=12)


plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Correlation between regional sales (heatmap).
regional_sales = df[['North America', 'Europe', 'Japan', 'Rest of World']]

correlation_matrix = regional_sales.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, linewidths=1)

plt.title('Correlation Between Regional Sales', fontsize=16)

plt.tight_layout()
plt.show()
```

```python
# Contribution of each genre to regional sales (stacked bar chart).

genre_region_sales = df.groupby('Genre')[['North America', 'Europe', 'Japan', 'Rest of World']].sum()

ax = genre_region_sales.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')

plt.title('Contribution of Each Genre to Regional Sales', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Sales (Millions)', fontsize=12)

plt.xticks(rotation=90, ha='right')

plt.tight_layout()
plt.show()

```
